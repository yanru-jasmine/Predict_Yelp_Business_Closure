## **Opening Day & Time**

```python
import re
business = open('yelp_academic_dataset_business.json', 'rb')
busi_data = [json.loads(line) for line in business]

def hour_count(string):
    if string == None:
        hour = 0.0
    else:
        pattern = r'(\d+):(\d+)-(\d+):(\d+)'
        h1, m1, h2, m2 = re.search(pattern,string).groups()
        hour = int(h2) - int(h1)
        hour += (int(m2) - int(m1))/60
        if int(h1) >= int(h2):
            hour += 24
    return hour

def day_count(dictionary):
    if dictionary == None:
        day = 0
    else:
        day = len(dictionary)
    return day

def hour_sum(dictionary):
    if dictionary == None:
        return 0.0
    else:
        hour_list = []
        for item in dictionary.values():
            hour_list.append(hour_count(item))
        return sum(hour_list)

def add_hour_attributes(data_list):
    for item in data_list:
        item['day_per_week'] = day_count(item['hours'])
        item['hour_per_week'] = hour_sum(item['hours'])
        try:
            item['hour_per_day'] = hour_sum(item['hours']) / day_count(item['hours'])
        except:
            item['hour_per_day'] = 0.0

add_hour_attributes(busi_data)
busi_df = pd.DataFrame.from_dict(busi_data, orient='columns')
```

```python
data_attribute = pd.read_csv('api_data.csv')
#we use yelp api to get raw data of business and store them in api_data csv file, and get six extra features from the raw data
```

```python
Claim = []
Price = []
Delivery = []
Pickup = []
for index, row in data_attribute.iterrows():
    if row['Claim_Data'] == True:
        Claim.append(1)
    else:
        Claim.append(0)
    if row['Price_Rating'] is not np.NaN:
        Price.append(row['Price_Rating'])
    else:
        Price.append(np.NaN)
    if row['Transaction_Data'] is np.NaN:
        Delivery.append(0)
        Pickup.append(0)
    elif len(row['Transaction_Data']) > 8:
        Delivery.append(1)
        Pickup.append(1)
    elif len(row['Transaction_Data']) > 6:
        Delivery.append(1)
        Pickup.append(0)
    else:
        Delivery.append(1)
        Pickup.append(0)

data_attribute['Claim'] = Claim
data_attribute['Price'] = Price
data_attribute['Delivery'] = Delivery
data_attribute['Pickup'] = Pickup

data_attribute['Price'].fillna(1.78)
```
